# matplotlib
matplotlibは、Pythonでグラフの描画や画像の表示などを行う外部モジュールです。グラフを描くには、以下のようにmatplotlibのpyplotというモジュールを使います。pyplotのデータにはNumPyの配列を使いますので、NumPyもインポートします。

```python
import numpy as np
import matplotlib.pyplot as plt
```

## plot関数
配列xとyを作り、折れ線グラフ（以下の例では直線）を描画します。pyplotのplot関数を使い、x座標とy座標のデータをプロットします。そして、show関数でグラフを表示します。

```python
# 配列を作り、折れ線グラフを描画
x = np.array([1, 2, 3, 4, 5])
y = np.array([100, 200, 300, 400, 500])
plt.plot(x, y)            # xとyをプロット
plt.show()                # グラフを表示
```

では、yに少しノイズを加えてみましょう。ノイズを、randomモジュールのrandn関数によって生成します。randn関数は、平均0、分散1の正規分布の乱数を生成します。したがって、yには、xに正規乱数が足された数値が入ります。散布図を描くなら、plot関数の3つ目の引数を"o"とし、点で描画することを指定します。

```python
# 散布図
num = 20                                            # データの数
x = np.arange(num)                                  # num個の1次元配列を作る
y = [x1 + np.random.randn(1) for x1 in x]           # ノイズを加える
print(x, y)
plt.plot(x, y, "o")                                 # xとyをプロット
plt.show()                                          # グラフを表示
```

## 直線近似
その上で、x, yの近似直線を求めてみましょう。多項式係数を導出する関数としてpolyfitがあります。引数はデータ配列(x, y)と多項式の次数（整数型）です。戻り値は近似多項式の各項の係数です。以下の例では1次近似式を設定していますので、戻り値はその傾きと切片になります。plot関数の3つ目の引数"-"は線で描画します。4つ目のcolorは色を指定します。

```python
# 1次近似式
a, b = np.polyfit(x, y, 1)          # 1次近似式を導出
print(a, b)                         # 傾きと切片
y_reg = a * x + b                   # 傾きと切片を指定して線形式を作る
plt.plot(x, y, "o")                 # xとyを点でプロット
plt.plot(x, y_reg, '-', color="red")      # xとyを赤線で描画
plt.show()                          # グラフを表示
```

## 三角関数
次に、sin関数およびcos関数を描画してみましょう。

```python
# 三角関数を描画する
x = np.linspace(0, 10)            # 要素数はdefaultで50。
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
plt.show()
```

グラフに凡例をつけてみよう。凡例の表示にはlegend関数が必要です。オプションの引数locを設定しますと、表示位置を指定できます。

```python
# 凡例を表示
x = np.linspace(0, 10)            # 要素数はdefaultで50。
plt.plot(x, np.cos(x), color="red", label="cos")
plt.plot(x, np.sin(x), color="blue", label="sin")
plt.legend()
plt.show()
```

グラフにタイトルをつけ、軸に名前を設定します。タイトルを付けるにはtitle関数を使います。軸に名前を設定するにはx軸に対してxlabel関数を、y軸に対してylablel関数を使います。

```python
# タイトルと軸をつける
x = np.linspace(0, 10)            # 要素数はdefaultで50。
plt.plot(x, np.cos(x), color="red", label="cos")
plt.plot(x, np.sin(x), color="blue", label="sin")
plt.legend()
plt.title("cos & sine")           # タイトル
plt.xlabel("x-axis")              # x軸
plt.ylabel("y-axis")              # y軸
plt.show()
```

## scatter関数
先ほど、散布図を描くためにplot関数を使い、オプションの引数に"o"を指定しました。散布図を描くにはscatter関数も使えます。以下のコードの中にあるrandomモジュールのrand関数は、0.0以上1.0未満の一様乱数を発生します。

```python
# 散布図にscatter関数を使う
x = np.random.rand(50)
y = np.random.rand(50)

# 散布図の描画
plt.scatter(x, y)
plt.show()
```

scatter関数のオプション引数で、散布図の点のスタイルを変更してみましょう。オプション引数cは点の色、sは点の大きさ、alphaは点の透明度、edgecolorsは点の周囲の線の色を指定します。

```python
# scatter関数のオプション引数
plt.scatter(x, y, c="purple", s=500, alpha=0.5, edgecolors="red")
plt.show()
```

オプション引数に変数を指定することもできます。例えば、色の引数cに変数yを代入すれば、yの値によって色を変化させることができます。これに伴ってカラーマップを使います。カラーマップは色を対応付けた一連の値のことで、一般にはRGB値との関係が定義されています。ここでは、引数cmapに予め用意されている"Blues"（青系の色）を使うことにします。

```python
# scatter関数のオプション引数
plt.scatter(x, y, c=y, s=500, cmap="Blues")
plt.show()
```

次に、グリッドとカラーバーを表示してみましょう。グリッドを表示するには、show関数の前にgrid関数を記述します。カラーバーはカラースケールを示すもので、show関数の前にcolorbar関数を記述します。

```python
# scatter関数のオプション引数
plt.scatter(x, y, c=y, s=500, cmap="Blues")
plt.grid()
plt.colorbar()
plt.show()
```

## グラフを画像ファイルとして保存する
matplotlibで作成したグラフをファイルとして保存するには、マウス・カーソルを右クリックして、ポップアップしたメニューで「名前を付けて画像を保存」とか、「画像をコピー」して別アプリで「貼り付け」とか、すれば可能です。ただ、matplotlibモジュールにも、作成したグラフを保存する機能があります。画像ファイルとして直接的に出力する方法を紹介しましょう。savefig関数を使います。ただし、ここではGoogle Colaboratoryを使っていますから、自身のコンピュータにファイルを保存するためのfilesモジュールが必要です。

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, c=y, s=500, cmap="Blues")
plt.xlabel("x-axis")              # x軸
plt.ylabel("y-axis")              # y軸
plt.grid()
plt.colorbar()

plt.savefig("scatter01.png")            # プロットしたグラフをファイルに保存する
files.download("scatter01.png")         # Google Colaboratoryでファイルを保存する
plt.show()
```

画像ファイルとして対応しているファイル形式は、「eps, jpeg, jpg, pdf, pgf, png, ps, raw, rgba, svg, svgz, tif, tiff」です。ファイル形式は基本的にその拡張子によって自動判別されるのですが、オプションformatで明示することができます。また、ラスタ表現の（画素の集まりで表現された）画像でしたら、オプションdpiを指定することでその解像度を指定できます。

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

x = np.arange(-7, 7, 0.1)
y = np.sin(x)
plt.plot(x, y)

plt.savefig("sin01.jpg", format="jpeg", dpi=400)          # プロットしたグラフをファイルに保存する
files.download("sin01.jpg")                               # Google Colaboratoryでファイルを保存する
plt.show()
```
